```python
import cv2
import numpy as np
import matplotlib.pyplot as plt
from google.colab.patches import cv2_imshow
from google.colab import files
from PIL import Image, ImageFilter, ImageEnhance
import io
```

```python
def upload_image():
    uploaded = files.upload()
    for filename in uploaded.keys():
        image = Image.open(io.BytesIO(uploaded[filename]))
        return image
    return None
```

```python
def apply_filters(image):
    filters = {
        "Original": image,
        "Grayscale": image.convert("L"),
        "Blur": image.filter(ImageFilter.BLUR),
        "Sharpen": image.filter(ImageFilter.SHARPEN),
        "Edge Enhance": image.filter(ImageFilter.EDGE_ENHANCE),
        "Contour": image.filter(ImageFilter.CONTOUR),
        "Emboss": image.filter(ImageFilter.EMBOSS),
        "Color Enhance": ImageEnhance.Color(image).enhance(2.0)
    }
    return filters
```

```python
def display_images(filters):
    plt.figure(figsize=(12, 8))
    for i, (name, img) in enumerate(filters.items()):
        plt.subplot(2, 4, i+1)
        plt.imshow(img, cmap='gray' if img.mode == 'L' else None)
        plt.title(name)
        plt.axis("off")
    plt.show()
print("Please upload an image")
image = upload_image()
```

```python
if image:
    filters = apply_filters(image)
    display_images(filters)
else:
    print("No image uploaded.")
```
